```python
import openpyxl as op
```

```python
load_data = op.load_workbook(r"dga_dataset.xlsx")
dataset = load_data.active
```

```python
domain = []
dga = []
for i in range(2, 10001):
    if(dataset.cell(row = i, column = 1).value == 'dga'):
        domain.append(dataset.cell(row = i, column = 3).value)
    else:
        dga.append(dataset.cell(row = i, column = 3).value)
```

```python
print(domain[0])
print(dga[0])
```

```python
# n - gram : n 개의 연속적인 단어의 나열
# n 지정

n = 4
```

```python
def n_gram(P, Q): # brute-force에서 차용
    count = 0
    for i in range(len(P) - n + 1):
        substringP = P[i : i + n]
        
        for j in range(len(Q) - n + 1):
            substringQ = Q[j : j + n]
            
            if(substringP == substringQ):
                count += 1 # 같은 부분이 존재하면

    if(i == len(P) - n - 1):
        return 0 
    
    return count #겹치는 부분의 갯수를 계산한다
```

```python
def Similarity(P, Q):
    ngramP = len(P) - n + 1
    ngramQ = len(Q) - n + 1
    
    min_value = 0
    
    if(ngramP > ngramQ):
        min_value = ngramQ
    else:
        min_value = ngramP
    
    count = n_gram(P, Q)
    
    result = count / min_value
    
    return result
```

```python
print(len(domain))
print(len(dga))
```

```python
# 도메인 넣기

result = []

for i in range(len(domain)):
    for j in range(len(dga)):
        if(Similarity(domain[i], dga[j]) >= 0.5):
            result.append([domain[i], dga[j], Similarity(domain[i], dga[j])])
```

```python
n = 4
```

```python
def n_gram(P, Q):
    count = 0
    for i in range(len(P) - n + 1):
        substringP = P[i : i + n]
        
        for j in range(len(Q) - n + 1):
            substringQ = Q[j : j + n]
            
            if(substringP == substringQ):
                count += 1
    
    return count
```

```python
def Similarity(P, Q):
    ngramP = len(P) - n + 1
    ngramQ = len(Q) - n + 1
    
    min_value = 0
    
    if(ngramP > ngramQ):
        min_value = ngramQ
    else:
        min_value = ngramP
    
    count = n_gram(P, Q)
    
    result = count / min_value
    
    return result
```

```python
#정상 도메인과 test data 비교

result = []

for i in range(len(domain)):#len(domain)
    print(i, end = " ")
    for j in range(1, 1000): #len(data) 9999부터 Nonetype
        SIMILAR = Similarity(domain[i], data[j])
        if(SIMILAR >= 0.7):
            result.append([domain[i], j, data[j], SIMILAR])
```
